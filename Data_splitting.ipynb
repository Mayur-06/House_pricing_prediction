```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#we directly use this which was saved in previous notebook (Data_Processing.ipynb)
dataDf = pd.read_csv('data.csv')

#actually you don't need to create csv files for everything and load and dumb from one notebook to another....
#I have done this to just separately show the process and steps in ML model building. You can run the whole code of project in one notebook also
#but if you do it will be a good habit. Your codes will be easier to understand be organized
```

```python
#splitting the dataset into data and target
x= dataDf.drop(['price'], axis =1)
y=dataDf['price']
```

```python
#splitting the data into training and testing by data I meant the one which includes the target value also. It is important to split target values also
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2, random_state=22)

#to use in another notebook
# Save the datasets to a CSV file format like before
pd.DataFrame(X_train).to_csv('X_train.csv', index=False)
pd.DataFrame(X_test).to_csv('X_test.csv', index=False)
pd.DataFrame(Y_train).to_csv('Y_train.csv', index=False)
pd.DataFrame(Y_test).to_csv('Y_test.csv', index=False)
```
